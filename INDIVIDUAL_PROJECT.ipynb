 Create a new empty DataFrame called returns. This dataframe
will contain the returns for each stock. Returns are obtained by 
$$r_t=\frac{p_t-p_{t-1}}{p_{t-1}}=\frac{p_t}{p_{t-1}}-1$$

```python
returns = pd.DataFrame()
```

 Use pandas pct_change() method on the Close column to create a column
 representing this return value and store that result to the returns. 

```python
for column in stocks_df.columns.levels[0]:  
    stock_close = stocks_df[column]['Close']
    stock_returns = stock_close.pct_change()  
    returns[column, 'Returns'] = stock_returns 
```

```python
for tick in tickers:
    returns[tick+' Return'] = stocks_df[tick]['Close'].pct_change()
returns.head()
```

# Moving Average

Plot the 30-day moving averages for JPM CHASE and Morgan Stanley stocks for the first 600 days. Use will need to use the individual daframe that you have create for each stock in Part I.

```python
import matplotlib.pyplot as plt
import seaborn as sns
sns.set_style('whitegrid')
%matplotlib inline
```

```python
plt.figure(figsize=(12,6))
stocks_df['JPM']['Close'].loc['2006-01-01':'2008-07-01'].rolling(window=30).mean().plot(label='30 Day Avg')
stocks_df['JPM']['Close'].loc['2006-01-01':'2008-07-01'].plot(label='JPM CLOSE')
stocks_df['MS']['Close'].loc['2006-01-01':'2008-07-01'].rolling(window=30).mean().plot(label='30 Day Avg')
stocks_df['MS']['Close'].loc['2006-01-01':'2008-07-01'].plot(label='MS CLOSE')
plt.legend()
```
